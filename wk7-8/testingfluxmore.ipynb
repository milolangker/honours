```python
import matplotlib.pyplot as plt
from scipy.special import jv
import jax.numpy as np
```

```python
differences = np.linspace(0,2*np.pi,100)

plt.plot(differences/np.pi, np.abs(jv(0, differences/4))**4)
plt.plot(differences/np.pi, np.abs(jv(1, differences/4))**4)
plt.plot(differences/np.pi, (jv(1, differences/4)/jv(0,differences/4))**4)
plt.xlim(0,1)
plt.ylim(0,1)
plt.show()
```
